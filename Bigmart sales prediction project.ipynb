# Data Preprocessing

## Handling Outliers

```python
data.describe()
```

```python
Q1 = np.percentile(data['Item_Outlet_Sales'],25)
Q3 = np.percentile(data['Item_Outlet_Sales'],75)
IQR = Q3-Q1

lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
```

```python
data = data[(data['Item_Outlet_Sales']>lower_limit) & (data['Item_Outlet_Sales']<upper_limit)]
```

```python
data.shape
```

## Handling Missing Values

```python
data.isnull().sum()
```

We will classify items according to item identifier feature

```python
data['Item_Identifier'].unique()
```

```python
data['Item_Identifier'] = data['Item_Identifier'].apply(lambda x: x[0:2])
data['Item_Identifier'] = data['Item_Identifier'].replace({'FD':'Food', 'NC':'Non-Consumable', 'DR':'Drinks'})
```

```python
data.head()
```

Missing item weights will be filled by the median value by their respective class

```python
item_weight_median = data.pivot_table(values='Item_Weight', columns='Item_Identifier', aggfunc=(lambda x: x.median()))
```

```python
item_weight_median
```

```python
missing_item_weight = data['Item_Weight'].isnull()
```

```python
missing_item_weight
```

```python
data.loc[missing_item_weight, 'Item_Weight'] = data.loc[missing_item_weight, 'Item_Identifier'].apply(lambda x: item_weight_median[x])
```

```python
data.isnull().sum()
```

Missing outlet size will be filled by the mode of their respective outlet type

```python
outlet_size_mode = data.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()))
```

```python
outlet_size_mode
```

```python
missing_values = data['Outlet_Size'].isnull()
```

```python
missing_values
```

```python
data.loc[missing_values, 'Outlet_Size'] = data.loc[missing_values, 'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
```

```python
data.isnull().sum()
```

```python
data['Item_Fat_Content'].unique()
```

```python
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace({'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'})
```

```python
data['Item_Fat_Content'].unique()
```

We will create an extra feature Outlet_Years according to how old these are

```python
data['Outlet_Years'] = 2021 - data['Outlet_Establishment_Year']
```

```python
data.head()
```

We will replace zeros in Item_Visibility column by the mean value

```python
len(data[data['Item_Visibility'] == 0])
```

```python
data['Item_Visibility'] = data['Item_Visibility'].replace(0, data['Item_Visibility'].mean())
```

```python
len(data[data['Item_Visibility'] == 0])
```

```python
data['Item_Type'].value_counts()
```

```python
data.head()
```

## Label Encoding

```python
data = data.drop(['Outlet_Identifier', 'Item_Type'], axis=1)
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.head()
```

```python
data = data.drop('Outlet_Establishment_Year', axis=1)
```

## Train-Test split

```python
X = data.drop('Item_Outlet_Sales', axis=1)
y = data['Item_Outlet_Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

# Model Building

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('MSE:', mean_squared_error(y_test, rfr.predict(X_test)))
```

```python
print('RMSE:', np.sqrt(mean_squared_error(y_test, rfr.predict(X_test))))
```

```python
print('r2 Score:', r2_score(y_test, rfr.predict(X_test)))
```

```python
plt.scatter(y_test, rfr.predict(X_test))
plt.plot(y_test, y_test, 'r')
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]

parameters = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
                'bootstrap':[True, False]}

print(parameters)
```

```python
ran_cv = RandomizedSearchCV(estimator=rfr, param_distributions=parameters, cv=5, n_jobs=-1, scoring='r2', verbose=2)
```

```python
ran_cv.fit(X_train, y_train)
```

```python
ran_cv.best_estimator_
```

```python
ran_cv.best_params_
```

```python
r2_score(y_test, ran_cv.predict(X_test))
```

```python
np.sqrt(mean_squared_error(y_test, ran_cv.predict(X_test)))
```
